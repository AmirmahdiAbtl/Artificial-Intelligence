```python
import numpy as np
import math
import random
```

```python
file = open('input.txt','r')
size = int(file.readline())
hunter = []
hunt = []
file_size = 0
for line in file:
    strings = ''
    for i in line:
        if i != " ":
            strings = strings + i
        if i == " ":
            hunter.append(int(strings))
            strings = ''
        if i == "\n":
            hunt.append(int(strings))
            strings = ''
            file_size += 1
```

```python
n = size
population_size = 30
max_pop_size = 50
crossover_coeff = 0.8
mutation_coeff = 0.4
max_iteration = 500
num_crossover = round(population_size*crossover_coeff)
num_mutation = round(population_size*mutation_coeff)
total = population_size + num_crossover + num_mutation

population = []
object_values = []
best_objective = 0
best_chromosome = np.zeros(n)
```

```python
def objective(column):
    q = 0
    for i in range(n):
        k = n - 1
        j = i + 1
        for j in range(n):
            if i < j:
                for z in range(file_size):
                    if hunter[z] == column[i] and hunt[z] == column[j]:
                        k = k - 1
        q = q + k
    return q
```

```python
while len(population) < population_size:
    solution = random.sample(range(0,n),n)
    for i in range(len(solution)):
        solution[i] = solution[i] + 1
    population.append(solution)
    object_values.append(objective(solution))
```

```python
def rating(x,point):
    rate = 0
    for i in range(n):
        if i != point:
            for j in range(file_size):
                if hunter[j] == x[point] and hunt[j] == x[i]:
                    rate = rate + 1
    return rate
```

```python
iteration = 0
while iteration < max_iteration:
    summation = sum(object_values)
    pr = []
    cumulative_pr = []
    for i in range (population_size):
        pr.append(1-object_values[i]/summation)
    cumulative_pr.append(pr[0])
    for i in range(1, population_size - 1):
        temp = cumulative_pr[i-1] + pr[i]
        cumulative_pr.append(temp)
    cumulative_pr.append(1)
    for i in range(0, int(num_crossover), 2):
        p1 = 0
        temp = np.random.rand()
        while cumulative_pr[p1] < temp:
            p1 = p1 + 1
        p2 = p1
        while p1 == p2:
            temp = np.random.rand()
            p = 0
            while cumulative_pr[p] < temp:
                p = p + 1
            p2 = p
        parent1 = population[p1]
        parent2 = population[p2]
        rand1 = np.random.randint(n)
        rand2 = np.random.randint(n)
        while rand1 == rand2:
            rand2 = np.random.randint(n) 
        A_rate1 = rating(parent1,rand1)
        B_rate1 = rating(parent2,rand1)
        A_rate2 = rating(parent1,rand2)
        B_rate2 = rating(parent2,rand2)
        child1 = parent1.copy()
        child2 = parent2.copy()       
        if (A_rate1 >= B_rate1):
            temp = child1[rand1]
            child1[rand1] = child2[rand1]
            for i in range(len(child1)):
                if i!=rand1 and child1[i] == child2[rand1]:
                    child1[i] = temp
        if (A_rate2 > B_rate2):
            temp = child1[rand2]
            child1[rand2] = child2[rand2]
            for i in range(len(child1)):
                if i!= rand2 and child1[i] == child1[rand2]:
                    child1[i] = temp
        child2 = child1.copy()
        rand1 = np.random.randint(n)
        rand2 = np.random.randint(n)
        while rand1 == rand2:
            rand2 = np.random.randint(n)
        A_rate1 = rating(parent1,rand1)
        B_rate1 = rating(parent2,rand1)
        A_rate2 = rating(parent1,rand2)
        B_rate2 = rating(parent2,rand2)
        if (B_rate1 >= A_rate1):
            temp = child2[rand1]
            child2[rand1] = child1[rand1]
            for i in range(len(child2)):
                if i!=rand1 and child2[i] == child1[rand1]:
                    child2[i] = temp
        if (B_rate2 > A_rate2):
            temp = child2[rand2]
            child2[rand2] = child1[rand2]
            for i in range(len(child2)):
                if i!= rand2 and child2[i] == child1[rand2]:
                    child2[i] = temp
        population.append(child1)
        population.append(child2)
        object_values.append(objective(child1))
        object_values.append(objective(child2))
    for i in range(num_mutation):
        nth_population = population_size + np.random.randint(num_crossover)
        mutated = population[nth_population].copy()
        mth_index = np.random.randint(n)
        random_value = np.random.randint(n)
        if random_value == mth_index:
            while random_value == mutated[mth_index]:
                random_value = np.random.randint(n)
        temp = mutated[mth_index]
        mutated[mth_index] = random_value
        for i in range(n):
            if i != mth_index and mutated[i] == random_value:
                mutated[i] = temp
        population.append(mutated)
        object_values.append(objective(mutated))
    best_objective = max(object_values)
    best_arg = np.argmax(object_values)
    best_chromosome = population[best_arg]
    if len(population) > max_pop_size:
        temp_population = []
        temp_objective = []
        args = np.argsort(object_values)
        for i in range(max_pop_size):
            t = i
            temp_population.append(population[args[t]])
            temp_objective.append(object_values[args[t]])
        population = temp_population
        object_values = temp_objective
        population_size = max_pop_size
    print(best_objective)
    iteration = iteration + 1   
print(best_chromosome)
print(best_objective)
```
