# Solving the Hunter problem with simulating annealing 

### How simulating annealing algorithm work:
* at first we should finding the objective function
* then we find a predictive neighbour of a random index
* then we should check objective function of this two data sest
* if the neighbour has better objective function we should swap the solution set by neighbour set
* and we continuous until the T value become zero

```python
import numpy as np
import math
import random
```

#### In this block i try to read a file and split it 
* size : size of input graph
* first [] : hunter data set
* second [] : hunt data set
* file_size : how many line does my input file has

```python
file = open('input.txt','r')
size = int(file.readline())
first = []
second = []
file_size = 0
for line in file:
    strings = ''
    for i in line:
        if i != " ":
            strings = strings + i
        if i == " ":
            first.append(int(strings))
            strings = ''
        if i == "\n":
            second.append(int(strings))
            strings = ''
            file_size += 1
```

#### in this block i try to set the essentioal variable

```python
n = size
T = 100000
t_change = 0.9999
```

### objective funciton:

* i check a index with another index and if i found the index which it isn't in the satisfaction situation then decrease the objective value of that index
* for example if we have 7 in the fourth index and have 15 in first index then if we know the No.7 animal can hunt No.15 i decrease the objective value of this data set 

* [15, 2, 1, 8, 7, 4, 3, 10, 19, 18, 17, 16, 20, 11, 6, 9, 12, 13, 14, 5] == objective value = 345 but after one step it decrease to 344

<h3 style="color: red">Note : </h3>
<p>[380 - 326] range</p>

```python
def objective(column):
    q = 0
    for i in range(n):
        k = n - 1
        j = i + 1
        for j in range(n):
            if i < j:
                for z in range(file_size):
                    if first[z] == column[i] and second[z] == column[j]:
                        k = k - 1
        q = q + k
    return q
```

### create a random list in [0 - 20] range

```python
solution = random.sample(range(0,n),n)
for i in range(len(solution)):
    solution[i] = solution[i] + 1
fitness = objective(solution)
```

### SwapPositions is a function for swapping two value of the list

```python
def swapPositions(list, pos1, pos2):
    list[pos1], list[pos2] = list[pos2], list[pos1]
    return list
```

# Main block

* at first copy the random list to neighbour variable
* then in the while loop we continious until T is become zero
* in this program i considered two variable each variable have a random value 
* first_predict , second_predict : random index for checking with each other and make neighbour 
   * for example first_predict, second_predict = 5, 8 : check the 5 index and 8 index
* Check if we have a value in first list and it has a same value with first_predict index of neighbour list and the second list we have a value which has a same value with second_predict index of neighbour list
* after that we found the objective value of the new data set then we calculated the delta value
* we should solve a fitness - fit because this problem solved with minimizing solution
* then we check delta value if its larger than 0 we can set solution variable with neighbour value and the fitness with fit
* but if its smaller than 0 we select this neighbour with pr Possibility 


```python
neighbour = solution.copy()
while T > 0:
    first_predict, second_predict = random.randint(0, n - 1), random.randint(0, n - 1)
    prev_T = T
    happend = 0
    for j in range(file_size):
        if first[j] == neighbour[first_predict] and second[j] == neighbour[second_predict]:
            if first_predict > second_predict:
                swapPositions(neighbour, first_predict, second_predict)
                happend = 1
        elif first[j] == neighbour[second_predict] and second[j] == neighbour[first_predict]:
            if first_predict < second_predict:
                swapPositions(neighbour, second_predict, first_predict)
                happend = 1

    fit = objective(neighbour)
    delta = fitness - fit
    if delta >= 0:
        solution = neighbour
        fitness = fit
    else:
        pr = math.exp(delta / T)
        if pr >= 0.999:
            solution = neighbour
            fitness = fit
    print(fitness)
    T = int(T * t_change)
        # print(T)
print (solution)
print(fitness)
```
